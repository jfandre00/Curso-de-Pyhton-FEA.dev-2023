### Qual  a proporo de passageiros que sobreviveram e no sobreviveram?


Para responder a esta pergunta, voc pode criar um grfico de pizza com as contagens de passageiros sobreviventes e no sobreviventes.

No contexto do conjunto de dados do Titanic, a expresso se refere  contagem do nmero de passageiros que sobreviveram e o nmero de passageiros que no sobreviveram ao desastre. A coluna "Sobreviveu" no conjunto de dados  possui valores binrios, onde:

**1 representa que o passageiro sobreviveu.**

**0 representa que o passageiro no sobreviveu (ou seja, faleceu).**


A ideia  calcular a contagem ou frequncia desses dois valores na coluna "Sobreviveu" para determinar quantos passageiros sobreviveram e quantos no sobreviveram.

```python
### seu cdigo comea aqui##

import matplotlib.pyplot as plt

# Calcule as contagens de sobreviventes e no sobreviventes
contagem_sobreviventes = df['Sobreviveu'].value_counts()
print(contagem_sobreviventes)
#contagem_mortos=df['Sobreviveu'].value_counts()[0]
#print(contagem_sobreviventes)
#print(contagem_mortos)

plt.pie(contagem_sobreviventes, labels= ["No Sobreviveu", "Sobreviveu"], autopct='%1.1f%%')
plt.title('Grfico de Sobreviventes')
plt.show()
```

**curiosidade: oq siginifica autopct='%1.1f%%**
    
O parmetro autopct em um grfico de pizza (pie chart) do Matplotlib  usado para formatar a exibio das porcentagens das fatias no grfico. Ele aceita uma string de formatao que controla como as porcentagens so exibidas.

%1.1f%%: Vamos dividi-lo em partes:
%1.1f: Este  um especificador de formato que indica que queremos exibir um nmero de ponto flutuante (float) com uma casa decimal.
%%: Este  usado para exibir o smbolo de porcentagem "%" literalmente.
Portanto, %1.1f%% significa que estamos formatando o nmero de ponto flutuante com uma casa decimal e, em seguida, adicionamos o smbolo de porcentagem "%" no final.

Por exemplo, se a porcentagem da fatia for 25.4%, autopct='%1.1f%%' far com que ela seja exibida como "25.4%". Isso ajuda a fornecer uma exibio mais precisa das porcentagens no grfico de pizza. Voc pode ajustar a formatao conforme necessrio para atender s suas preferncias.

### Qual  a proporo de sobreviventes por sexo?

Passo a Passo:

**Calcule a contagem de sobreviventes por sexo. (utilize groupby)**

Neste passo, voc est agrupando os dados por sexo (gnero) e, em seguida, calculando a contagem de sobreviventes para cada grupo. Isso significa que voc est contando quantos passageiros sobreviveram para cada categoria de sexo (masculino e feminino).


**Calcule a proporo de sobreviventes para cada sexo**

Agora, voc est calculando a proporo de sobreviventes para cada categoria de sexo. Isso  feito dividindo o nmero de sobreviventes para cada sexo pelo total de sobreviventes.

**Exiba a proporo de sobreviventes por sexo**

Neste passo, voc est simplesmente exibindo a proporo de sobreviventes por sexo. Isso mostrar a proporo de passageiros sobreviventes do sexo masculino e feminino em relao ao total de sobreviventes.


**Crie um grfico de barras para representar a proporo de sobreviventes por sexo**


```python
df_sobreviventes = df.drop(df[df['Sobreviveu'] == 0].index)
sobrev = df_sobreviventes[['Sobreviveu', 'Sexo']].groupby('Sexo').count()
total_sobreviventes = sobrev['Sobreviveu'][0] + sobrev['Sobreviveu'][1]
prop_male = 100 * (sobrev['Sobreviveu'][1] / total_sobreviventes)
prop_female = 100 * (sobrev['Sobreviveu'][0] / total_sobreviventes)
plt.bar(['male', 'female'], [prop_male, prop_female])
plt.title('Grfico de Sobreviventes')
plt.ylabel('Percentual de Sobreviventes')
plt.show()
```
