```python
#https://www.kaggle.com/datasets/brenda89/fifa-world-cup-2022
import pandas as pd

df = pd.read_csv('international_matches.csv')
df.describe()
```

```python
df.head(5)
df = df.drop_duplicates()
#vamos remover os duplicados antes de remover a coluna 'date', pois pode acontecer de ter tido algum jogo com o mesmo
#resultado entre os times.
display(df)
```

```python
df = df.drop(columns=['date', 'home_team_total_fifa_points', 'away_team_total_fifa_points', 'shoot_out', 'home_team_goalkeeper_score', 'away_team_goalkeeper_score', 'home_team_mean_defense_score', 'home_team_mean_offense_score', 'home_team_mean_midfield_score', 'away_team_mean_defense_score', 'away_team_mean_offense_score', 'away_team_mean_midfield_score'])
display(df)
```

```python
df = df.drop(columns=['tournament', 'city', 'country'])
display(df)
```

```python
df = df.rename(columns={'home_team': 'casa', 'away_team': 'visitante', 'home_team_continent': 'casa_continente', 'away_team_continent': 'visitante_continente'})
display(df)
```

```python
df = df.rename(columns={'home_team_fifa_rank': 'rank_casa', 'away_team_fifa_rank': 'rank_visitante', 'home_team_score': 'gols_casa', 'away_team_score': 'gols_visitante'})
display(df)
```

```python
df = df.rename(columns={'neutral_location': 'campo_neutro', 'home_team_result': 'resultado_casa'})
display(df)
```
