```python
#Groupby - métodos úteis

vendas_mercado = pd.read_csv("supermarket_sales.csv")
vendas_mercado.info()
```

```python
display(vendas_mercado)
```

```python
#o método .groupby() organiza o data frame em grupos baseado no resultado de uma ou mais colunas

vendas_mercado.groupby("Payment")["Quantity"].count()

# agrupamos por método de pagamento e queremos saber a quantidade vendida por tipo de método
```

```python
#Segue um exemplo
vendas_mercado.groupby("Product line")["Quantity"].count()
```

```python
#o método .groupby() pode agrupar dados com base em mais de uma coluna

vendas_mercado.groupby(["Product line","Gender"])["Quantity"].count()

#agrupamos por linha de produto e queremos saber a quantidade gasta por cada gênero
```

```python
#Da mesma forma, podemos querer saber o faturamento total por branch (cada uma das três cidades)
vendas_mercado.groupby("City")["Total"].sum()
```

```python
#Pivot table
#podemos construir uma pivot table a partir de:
# de um index (Index)
# colunas de interesse (Columns)
# Valores que queremos agregar (values)
# Função agregadora (aggfunc)

vendas_mercado.pivot_table(index="City", columns=["Product line", "Gender"], values="Total", aggfunc="sum")

#queremos saber a soma do faturamento por gêneros dentro de cada linha de produto, separado por cidade
```

```python
#Merge 

salarios_funcionarios = pd.read_excel("salario.xlsx")

display(salarios_funcionarios)
```

```python
# e demais informações pessoais em outra planilha

informacoes_pessoais = pd.read_excel("informacoes_pessoais.xlsx")
display(informacoes_pessoais)
```

```python
#Note que as planilhas possuem em comum o nome dos funcionários

pd.merge(salarios_funcionarios, informacoes_pessoais, on="Nome")
```

```python
# É também possível definir como os DataFrames serão unidos a partir de um parâmetro how:
# how="inner": Inclui apenas os valores correspondentes em ambos os DF´s
# how="left" : Inclui todas as linhas do DF da direita e apenas as linhas corresp. do DF da esquerda. Linhas não corresp. 
# são preenchidas com NaN
# how="right" : oposto do left
# how="outer"> inclui todas as linhas de ambos os DF´s. Linhas não correspondentes são preenchidas com NaN

#Exemplo -> Inner, inclui apenas valores correspondentes em ambos os DataFrames
pd.merge(salarios_funcionarios, informacoes_pessoais, on="Nome", how="inner")
```

```python
# exemplo -> outer, inclui todos os valores disponíveis em ambos os DataFrames

pd.merge(salarios_funcionarios, informacoes_pessoais, on= "Nome", how="outer")
```

```python
# Exemplo -> right. Inclui todos os valores do DF a direita, aqueles não correspondentes são preenchidos com NaN

pd.merge(salarios_funcionarios, informacoes_pessoais, on="Nome", how="right")
```

```python
# Left -> inclui todos os valores do DF a esquerda, aqueles não correspondentes são preenchidos com NaN

pd.merge(salarios_funcionarios, informacoes_pessoais, on="Nome", how="left")
```

```python
#Diff -> sendo muito útil para analisar a taxa de variação ou tendência de uma base de dados, a função retornará todas as 
#variações entre uma linha e sua anterior para cada coluna do DataFrame. No caso de uma Series a função retornará apenas
#uma coluna

#PIB do Brasil
pib = pd.read_excel("pib_brasil.xlsx")
display(pib)
```

```python
#Exemplo

pib["PIB"].diff()
```

```python
#periods = n
#para periods = 1 (default) a função calcula a diferença entre a linha atual e a última linha, para periods=2 a função
#calcula a diferença entre a linha atual e a antepenúltima linha

pib["PIB"].diff(periods = 4)
```

```python
# pct_change -> opera de maneira similar ao diff(), porém ao inves de retornar valores absolutos , a função retorna a variação
#percentual entre  valores consecutivos
#o metodo nos auxilia a analisar o crescimento ou declinio relativo dos pontos de dados, especialmente em situações em que é
#importante compreender a taxa de mudança, tais como análises financeiras

# multiplicamos por 100 para termos valores percentuais
pib["PIB"].pct_change()*100
```

```python
#tb é possivel utilizar periods = n -> nesse caso (n=4),estaremos sempre comparando os mesmos trimestres

pib["PIB"].pct_change(periods=4)*100
```

```python
# Shift -> é util para analise de séries em função do tempo, calculo de diferencas entre valores deslocados e execução de vários
#tipos de transformações de dados, a função shift() é usada para deslocar os valores de uma série ou DF ao longo de um eixo
#específico, ou seja, move-los para cima ou para baixo nas posições do índice

pib_shift = pib.shift(periods=2)
display(pib_shift)
```

```python
# fillna()
# quando todos os valores são deslocados para baixo os valores do topo ficam vazios. Assim podemos definir um parâmetro 
# fillna(value=n) que preenche valores vazios criados pelo deslocamento

print(pib_shift.isnull().sum())
```

```python
# para alterar os valores para 0

pib_shift2 = pib_shift.fillna(value=0)
pib_shift2.head()
```

```python
# da mesma forma, pode se alterar para qualquer outro valor que desejar
pib_shift25 = pib_shift.fillna(value = 25)
pib_shift25.head()
```
